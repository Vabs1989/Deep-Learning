```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D
from keras.utils import to_categorical
from keras.preprocessing import image
from tqdm import tqdm
```

```python
train = pd.read_csv('E:/Diseases classification/Input/All_Training_70Patient_str_5.csv')    # reading the csv file
valid = pd.read_csv('E:/Diseases classification/Input/All_Validation_10Patient_str_5.csv')    # reading the csv file
test = pd.read_csv('E:/Diseases classification/Input/All_Testing_20Patient_str_5.csv')    # reading the csv file
train.head()      # printing first five rows of the file
```

```python
train.columns
```

```python
#img_dir = os.path.join('E:/Diseases classification/Input/resized/i256')
```

```python
train_image = []
for i in tqdm(range(train.shape[0])):
    img = np.load('E:/Diseases classification/Input/resized/i256/'+train['patientId'][i]+'.npy')
    img = image.img_to_array(img)
    #img = img/255
    train_image.append(img)
#X = np.array(train_image)
```

```python
valid_image = []
for i in tqdm(range(valid.shape[0])):
    img = np.load('E:/Diseases classification/Input/resized/i256/'+valid['patientId'][i]+'.npy')
    img = image.img_to_array(img)
    #img = img/255
    valid_image.append(img)
#X = np.array(train_image)
```

```python
test_image = []
for i in tqdm(range(test.shape[0])):
    img = np.load('E:/Diseases classification/Input/resized/i256/'+test['patientId'][i]+'.npy')
    img = image.img_to_array(img)
    #img = img/255
    test_image.append(img)
#X = np.array(train_image)
```

```python
X_train = np.array(train_image)
X_train.shape
```

```python
X_valid = np.array(valid_image)
X_valid.shape
```

```python
y_train = np.array(train.drop(['patientId', 'label', 'Set', 'age','sex', 'view'],axis=1))
y_train.shape
```

```python
y_valid = np.array(valid.drop(['patientId', 'label', 'Set', 'age','sex', 'view'],axis=1))
y_valid.shape
```

```python
y_test = np.array(test.drop(['patientId', 'label', 'Set', 'age','sex', 'view'],axis=1))
y_test.shape
```

```python
model = Sequential()
model.add(Conv2D(filters=16, kernel_size=(5, 5), activation="relu", input_shape=(256,256,1)))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
model.add(Conv2D(filters=32, kernel_size=(5, 5), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
model.add(Conv2D(filters=64, kernel_size=(5, 5), activation="relu"))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
model.add(Conv2D(filters=64, kernel_size=(5, 5), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(5, activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=10, validation_data=(X_valid, y_valid), batch_size=42)
```
